```python
# encoding the gender column
replace_dict = {'m' : 0,
                'f' : 1}

df['gender'] = df['gender'].replace(replace_dict)
df['gender'] = df['gender'].replace(replace_dict)
```

```python
df.head()
```

```python
# encoding the type column 
replace_dict = {'private' : 0,
                'National' : 1,
                'Municipal': 2,
                'county':3}

df['type'] = df['type'].replace(replace_dict)
df['type'] = df['type'].replace(replace_dict)
df['type'] = df['type'].replace(replace_dict)
df['type'] = df['type'].replace(replace_dict)
```

```python
df.head()
```

```python
X=df.drop(columns=['school status'], axis=1)
y=df['school status']
```

```python
X.head()
```

```python
df.columns
```

```python
y.head()
```

## train test split

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.30, random_state=42)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

## MLP classification

```python
from sklearn.neural_network import MLPClassifier

X=df.drop(columns=['school status'], axis=1)
y=df['school status']
clf = MLPClassifier(solver='lbfgs', alpha=1e-5,hidden_layer_sizes=(5, 2), random_state=1)
clf.fit(X, y)
```

```python
y_pred_model1 = clf.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,y_pred_model1)
```

```python
from sklearn.metrics import recall_score,make_scorer
recall_scorer = make_scorer(recall_score, pos_label = 'Unsuccessful graduation')

# Assuming 'Unsuccessful graduation' is the positive class
pos_label = 'Unsuccessful graduation'

recall = recall_score(y_test, y_pred_model1, pos_label=pos_label)

print(f'The recall for class {pos_label} is: {round(recall, 4)*100}%')
```

```python
from sklearn.metrics import f1_score

# Assuming 'Unsuccessful graduation' is the positive class
pos_label = 'Unsuccessful graduation'

f1 = f1_score(y_test, y_pred_model1, pos_label=pos_label)

print(f'The F1 score for class {pos_label} is: {round(f1, 4)*100}%')
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.neural_network import MLPClassifier

# define the parameter grid to search over
param_grid = {
    'hidden_layer_sizes': [(5,), (10,), (5,2), (10,2)],
    'activation': ['relu', 'tanh', 'logistic'],
    'alpha': [0.0001, 0.001, 0.01]
}

# create an MLPClassifier object
mlp = MLPClassifier(max_iter=1000)

# create a GridSearchCV object and fit it to the data
grid_search = GridSearchCV(mlp, param_grid, cv=5,scoring= recall_scorer)
grid_search.fit(X_train, y_train)

# print the best hyperparameters and the corresponding accuracy score
print("Best hyperparameters: ", grid_search.best_params_)
print("Recall score: ", grid_search.best_score_)
```
